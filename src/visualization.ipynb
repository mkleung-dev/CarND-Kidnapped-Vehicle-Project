```python
fig = plt.figure(figsize=(16, 15))

fig.add_subplot(2, 1, 1)
plt.plot(particle_count, err_x, '-o')
plt.plot(particle_count, err_y, '-o')
plt.title('Error Study')
plt.xscale('log')
plt.ylabel('Error (m)')
plt.xlabel('Particle Count')
plt.legend(['X Error', 'Y Error'], loc='right')

fig.add_subplot(2, 1, 2)
plt.plot(particle_count, err_theta, '-o')
plt.title('Error Study')
plt.xscale('log')
plt.ylabel('Error (m)')
plt.xlabel('Particle Count')
plt.legend(['Theta Error'], loc='right')

plt.show()
```
